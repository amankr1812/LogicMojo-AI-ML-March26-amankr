Q8. Group-Based Insights (Customer Segmentation Logic)

Q13. Custom Feature Transformation using apply()

Context: Real ML requires custom logic transformation.

Dataset: Titanic

Tasks:

1. Create age_group:
Child / Adult / Senior
2. Use .apply()
3. Compute survival rate per group
4. Interpret:
Which segment has highest survival likelihood

```python
df = sns.load_dataset('titanic')

def categorize_age(age):
    if age < 18:
        return 'Child'
    elif age <= 60:
        return 'Adult'
    elif age > 60:
        return 'Senior'
    else:
        return 'Unknown' 

df['age_group'] = df['age'].apply(categorize_age)
df['age'].isna().sum()
#df.head()
survival_by_age = df.groupby('age_group')['survived'].mean()
print(survival_by_age)
```
